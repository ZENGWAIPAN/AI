```python
import pandas as pd
import statsmodels.stats.contingency_tables as tbl
```

```python
home = pd.read_excel("home_income.xlsx")
home.head()
```

```python
pd.crosstab(home.Ts9, home.O1)
```

```python
#考察不同收入級別的家庭轎車擁有率是否相同
table = tbl.Table(pd.crosstab(home.Ts9, home.O1))
```

```python
res = table.test_nominal_association()
print("卡方值",res.statistic)
print("自由度",res.df)
print("p值",res.pvalue) #格式化後的p值
```
