```python
# 構造數據集
import numpy as np
np.random.seed(666)
x = np.random.uniform(-3.0,3.0,size=100)
X = x.reshape((-1,1))
y = 0.5 * x + 3 + np.random.normal(0,1,size=100) # 原始數據的特征和標簽

# 可視化
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
```

```python
def ridge_regression(degree, alpha):
  return Pipeline([
      ("poly", PolynomialFeatures(degree=degree)),
      ("std", StandardScaler()), #特征標準化
      ("ridge", Ridge(alpha=alpha))
  ])
```

```python
pipe_line = ridge_regression(20,0.001)
pipe_line.fit(X,y) # 經過Pipeline操作
y_pred = pipe_line.predict(X)
plt.scatter(x,y)
plt.plot(np.sort(x), y_pred[np.argsort(x)], color='red') #預測值
plt.show()
```
