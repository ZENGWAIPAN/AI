```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
#數據加載，展示圖像
digits = datasets.load_digits()
images = digits.images # 所有圖像數據
plt.gray() #圖像灰化
plt.matshow(images[0]) #顯示第一幅圖像
```

```python
X = digits.data #樣本特征
y = digits.target #樣本標籤

# 拆分數據集
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.4,random_state=666)
```

```python
# 開始交叉驗證
from sklearn.model_selection import cross_val_score

best_k, best_p, best_score = 0,0,0

for k in range(2,11): # 外層循環搜索k
  for p in range(1,6): #內層循環搜索p
    knn = KNeighborsClassifier(weights="distance",n_neighbors=k,p=p)
    scores = cross_val_score(knn,X_train,y_train,cv=3) # 3折交叉驗證
    score = np.mean(scores) #當前這一組超參數在驗證集上的平均分
    if score > best_score:
      best_k, best_p, best_score = k,p,score

print(f"{best_k=}")
print(f"{best_p=}")
print(f"{best_score=}")
```

```python
# 使用調好的超參數進行訓練和測試
best_knn = KNeighborsClassifier(weights="distance", n_neighbors=best_k, p=best_p)
best_knn.fit(X_train, y_train)
best_knn.score(X_test, y_test)
```
