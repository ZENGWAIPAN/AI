```python
import pandas as pd
```

```python
df = pd.read_excel("/content/IT行业收入表.xlsx")
df.head()
```

```python
X = df[["工龄"]]
y = df[["薪水"]]
```

```python
!pip install wget
import wget
wget.download("https://github.com/GrandmaCan/ML/raw/main/Resgression/ChineseFont.ttf")
```

```python
# 绘制散点图
from matplotlib import pyplot as plt
from matplotlib.font_manager import fontManager
import matplotlib as mlp

fontManager.addfont("ChineseFont.ttf")
mlp.rc('font', family="ChineseFont")

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.scatter(X,y)
plt.xlabel('工龄')
plt.ylabel('薪水')
plt.show()
```

```python
#模型搭建
from sklearn.linear_model import LinearRegression

regr = LinearRegression()
regr.fit(X,y) # 训练模型
```

```python
#模型可视化
plt.scatter(X,y)
plt.plot(X, regr.predict(X), color='red') # 线性回归模型的可视化
plt.xlabel('工龄')
plt.ylabel('薪水')
plt.show()
```

```python
regr.coef_ #系数
```

```python
regr.intercept_ #截距
```

```python
# 线性回归方程构造
print("系数w为："+str(regr.coef_[0]))
print("截距b为"+str(regr.intercept_))
```
