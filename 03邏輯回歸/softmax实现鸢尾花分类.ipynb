```python
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=666)
```

```python
# 如果是softmax回归，就要将multi_class参数设置为multinomial
sm = LogisticRegression(multi_class="multinomial")
sm.fit(X_train, y_train)
pred_probs = sm.predict_proba(X_test) # 在测试集上预测样本在每个分类的概率
pred_probs
```

```python
import numpy as np
np.sum(pred_probs[0]) # 针对每个样本预测出在每个分类上的概率
```

```python
sm.score(X_test, y_test)
```
