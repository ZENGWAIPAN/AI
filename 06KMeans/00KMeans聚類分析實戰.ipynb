```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(n_samples=1000,n_features=2,centers=[[-1,-1],[1,1],[2,2],[0,0]],cluster_std=[0.4,0.2,0.2,0.2],random_state=666)

plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
# KMeans 聚成2類觀察
from sklearn.cluster import KMeans

km = KMeans(n_clusters=4, random_state=666) #創建KMeans算法對象，設置聚成2類
km.fit(X) # 無監督學習，擬合的時候不需要樣本標簽
y_pred = km.predict(X) #預測
plt.scatter(X[:,0], X[:,1], c=y_pred) #預測為同一簇的樣本同顏色
plt.show()
```

```python
# 使用CH指標評價聚類效果（本題是在訓練樣本集上評價）
from sklearn.metrics import calinski_harabasz_score
calinski_harabasz_score(X,y_pred)
```
