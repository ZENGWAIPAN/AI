```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.read_excel("客户信息.xlsx")
data.head()
```

```python
from matplotlib import pyplot as plt
import matplotlib
a=sorted([f.name for f in matplotlib.font_manager.fontManager.ttflist])

for i in a:
    print(i)
```

```python
plt.rcParams["font.family"] = ["FZShuTi"]
plt.scatter(data["年龄(岁)"], data["收入(万元)"],c="green",marker="*")
plt.xlabel("年龄（岁）")
plt.ylabel("收入（万元）")
plt.show()
```

```python
# 模型训练
kms = KMeans(n_clusters=3, random_state=123) # n_cluster 分多少个类
kms.fit(data)
label = kms.labels_
print("聚类的结果：",label)
```

```python
# 建模效果可视化展示 shift alt down
plt.scatter(data[label==0].iloc[:,0],data[label==0].iloc[:,1],c="red",marker="o",
            label="class0")
plt.scatter(data[label==1].iloc[:,0],data[label==1].iloc[:,1],c="green",marker="*",
            label="class1")
plt.scatter(data[label==2].iloc[:,0],data[label==2].iloc[:,1],c="blue",marker="+",
            label="class2")

plt.xlabel("年龄（岁）")
plt.ylabel("收入（万元）")
plt.legend() # 设置图例
plt.show()
```
